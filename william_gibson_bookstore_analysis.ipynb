```python
import pandas as pd
import numpy as np
```

```python
books_data = {
    "Book": [
        "Neuromancer",
        "Count Zero",
        "Mona Lisa Overdrive",
        "Pattern Recognition",
        "Spook Country",
        "Zero History",
        "The Peripheral",
        "Agency"
    ],

    "Genre": [
        "Cyberpunk",
        "Cyberpunk",
        "Cyberpunk",
        "Sci-Fi",
        "Sci-Fi",
        "Sci-Fi",
        "Sci-Fi",
        "Sci-Fi"
    ],

    "Price (£)": [
        9.99,
        9.99,
        9.99,
        10.99,
        10.99,
        10.99,
        12.99,
        12.99
    ],

    "Cost (£)": [
        5.00,
        5.00,
        5.00,
        6.00,
        6.00,
        6.00,
        7.00,
        7.00
    ],

    "Popularity": [
        0.95,
        0.80,
        0.75,
        0.85,
        0.70,
        0.65,
        0.90,
        0.85
    ]
}

books_df = pd.DataFrame(books_data)

books_df
```

```python
companies_data = {

    "Company": [
        "Waterstones",
        "WHSmith"
    ],

    "Stores": [
        317,
        500
    ],

    "Customers_per_day_per_store": [
        350,
        250
    ],

    "Profit_margin": [
        0.18,
        0.12
    ],

    "Fiction_percent": [
        0.55,
        0.35
    ]
}

companies_df = pd.DataFrame(companies_data)

companies_df
```

```python
demographics_data = {

    "Age_Group": [
        "18-25",
        "26-35",
        "36-50",
        "51+"
    ],

    "Population_percent": [
        0.28,
        0.32,
        0.25,
        0.15
    ],

    "SciFi_interest": [
        0.70,
        0.65,
        0.50,
        0.30
    ]
}

demographics_df = pd.DataFrame(demographics_data)

demographics_df
```

```python
companies_df["Weekly_customers"] = (
    companies_df["Stores"] *
    companies_df["Customers_per_day_per_store"] *
    7
)

companies_df
```

```python
def estimate_buyers(company, popularity):

    buyers = (
        company["Weekly_customers"]
        * company["Fiction_percent"]
        * popularity
        * 0.03
    )

    return int(buyers)
```

```python
results = []

for _, company in companies_df.iterrows():

    for _, book in books_df.iterrows():

        buyers = estimate_buyers(company, book["Popularity"])

        revenue = buyers * book["Price (£)"]

        cost = buyers * book["Cost (£)"]

        profit = (revenue - cost) * company["Profit_margin"]

        yearly_profit = profit * 52

        results.append({

            "Company": company["Company"],
            "Book": book["Book"],
            "Weekly_buyers": buyers,
            "Weekly_profit (£)": round(profit, 2),
            "Yearly_profit (£)": round(yearly_profit, 2)

        })

results_df = pd.DataFrame(results)

results_df
```

```python
summary = results_df.groupby("Company").sum()

summary
```
