```python
import pandas as pd
import numpy as np
```

```python
companies = ["Pearson", "AI Books"]
companies
```

```python
bookstores = [
    "London",
    "Manchester",
    "Birmingham",
    "Leeds",
    "Liverpool",
    "Oxford",
    "Cambridge",
    "Bristol",
    "Brighton",
    "Nottingham"
]
```

```python
customers = [
    "Students",
    "Professionals",
    "Researchers",
    "General Readers",
    "Tech Enthusiasts"
]

age_groups = [
    "16-24",
    "25-34",
    "35-44",
    "45-54",
    "55+"
]
```

```python
logic = {

    "Pearson": {
        "price": 45,
        "profit_margin": 0.13
    },

    "AI Books": {
        "price": 25,
        "profit_margin": 0.22
    }
}
```

```python
data = []

for store in bookstores:
    
    for company in companies:
        
        weekly_customers = np.random.randint(500, 3000)
        
        price = logic[company]["price"]
        margin = logic[company]["profit_margin"]
        
        weekly_sales = weekly_customers * np.random.uniform(0.4, 0.7)
        
        weekly_revenue = weekly_sales * price
        weekly_profit = weekly_revenue * margin
        
        yearly_revenue = weekly_revenue * 52
        yearly_profit = weekly_profit * 52
        
        customer_type = np.random.choice(customers)
        age = np.random.choice(age_groups)
        
        data.append([
            store,
            company,
            customer_type,
            age,
            int(weekly_customers),
            int(weekly_sales),
            weekly_revenue,
            weekly_profit,
            yearly_revenue,
            yearly_profit
        ])
```

```python
columns = [
    "Store",
    "Company",
    "Customer_Type",
    "Age_Group",
    "Weekly_Customers",
    "Weekly_Sales",
    "Weekly_Revenue",
    "Weekly_Profit",
    "Yearly_Revenue",
    "Yearly_Profit"
]

df = pd.DataFrame(data, columns=columns)

df
```

```python
df.groupby("Company")[["Weekly_Profit", "Yearly_Profit"]].sum()
```

```python
df.to_csv("pearson_vs_ai_books_england.csv", index=False)
```
