```python
import pandas as pd
import numpy as np
import random
```

```python
companies = {

    "Waterstones": {
        "stores": 317,
        "avg_price": 10.99,
        "cost": 6.00,
        "weekly_sales_per_store": 80
    },

    "WHSmith": {
        "stores": 500,
        "avg_price": 9.99,
        "cost": 5.50,
        "weekly_sales_per_store": 55
    }

}
```

```python
books = [

    "A Game of Thrones",
    "A Clash of Kings",
    "A Storm of Swords",
    "A Feast for Crows",
    "A Dance with Dragons",
    "Fire and Blood"

]
```

```python
demographics = [

    "Teenagers",
    "Young Adults",
    "Adults",
    "Older Adults"

]

customer_types = [

    "Fantasy Fans",
    "TV Show Fans",
    "Collectors",
    "Gift Buyers",
    "Casual Readers"

]
```

```python
dataset = []

for company in companies:

    stores = companies[company]["stores"]
    price = companies[company]["avg_price"]
    cost = companies[company]["cost"]
    weekly_sales = companies[company]["weekly_sales_per_store"]

    for book in books:

        units_sold = stores * weekly_sales * random.uniform(0.7, 1.3)

        revenue = units_sold * price
        total_cost = units_sold * cost
        profit = revenue - total_cost

        dataset.append({

            "Company": company,
            "Book": book,
            "Units Sold Weekly": int(units_sold),
            "Revenue Weekly (£)": int(revenue),
            "Cost Weekly (£)": int(total_cost),
            "Profit Weekly (£)": int(profit),
            "Customer Type": random.choice(customer_types),
            "Demographic": random.choice(demographics)

        })

df = pd.DataFrame(dataset)

df
```

```python
weekly_profit = df.groupby("Company")["Profit Weekly (£)"].sum()

weekly_profit
```

```python
yearly_profit = weekly_profit * 52

yearly_profit
```
